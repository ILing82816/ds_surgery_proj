```python
import os
import pandas as pd
```

```python
data_prefix = 'D:/USA 2020 Spring/data mining'
```

```python
testdf = pd.read_csv( os.path.join(data_prefix, 'Python Lab/PYTHON_LAB_DF_TEST_2.csv'))
```

```python
testdf.head()
```

```python
testdf.info()
```

```python
testdf.dropna(subset = ["FEMALE"], inplace=True)
```

```python
for col in testdf:
    if sum(pd.isnull(testdf[col]))>len(testdf) * 0.95:
        testdf.drop([col],axis=1,inplace=True)
print('\nTest shape: {}'.format(testdf.shape)) 
```

```python
testdf = testdf.drop(["CAV_REC_SEX","CAV_REC_LANG","CAV_REC_AGE"], axis=1)
```

```python
for col in testdf:
    if testdf[col].dtype == "object":
        testdf[col].fillna(testdf[col].mode()[0], inplace=True)
    else:
        testdf[col].fillna(testdf[col].mean(), inplace=True)
```

```python
def age_simple(age):
    if age<=44:
        return "young"
    elif age>44 and age<=59:
        return "mid_age"
    elif age>59 and age<=74:
        return "old"
    elif age>74:
        return "older"
```

```python
testdf["age_people"] = testdf["AGE_ON_CONTACT_DATE"].apply(age_simple)
```

```python
cat_col = []
for col in testdf:
    if testdf[col].dtype == "object":
        cat_col.append(col)
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for column in cat_col:
    testdf[column] = lb.fit_transform(testdf[column])
```

```python
testdf.head()
```

```python
X_test = testdf.drop(["ID1","AGE_ON_CONTACT_DATE"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_test_vars = scaler.fit_transform(X_test)
```

```python
import pickle
with open(os.path.join(data_prefix, 'ds_surgery_proj/pre_lgbm_model.pckl'), 'rb') as fin:
        model = pickle.load(fin)
```

```python
testdf['risk_prob'] = model.predict(X_test_vars)
```
